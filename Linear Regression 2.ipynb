### Aggregate to get a count per hour/minute of the day across all trips

```python
reg_bikes = bikes[bikes['Member Type']=='Registered']
hours = reg_bikes.groupby([reg_bikes.hour_of_day, reg_bikes.start.dt.dayofweek]).agg('count')
hours['hour'] = hours.index
day_hour_count = hours.dur.unstack()
plt.figure(figsize=(20,10))
plt.plot(day_hour_count.index, day_hour_count[0])
plt.plot(day_hour_count.index, day_hour_count[1])
plt.plot(day_hour_count.index, day_hour_count[2])
# plt.plot(y.index, day_hour_count[3])
# plt.plot(y.index, day_hour_count[4])
# plt.plot(y.index, day_hour_count[5])
# plt.plot(y.index, day_hour_count[6])
```


## 2a. Create 3 models fit to `monday.hour_of_day` with varying polynomial degrees ( choose from `n=1,2,3,5,10,15`). (Repeat for `saturday` below)

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
hour = np.array(day_hour_count.index)
x = hour.reshape(-1, 1)
y = monday[0].fillna(0)
```

```python
poly_15 = PolynomialFeatures(degree = 15)
x_15 = poly_15.fit_transform(x.reshape(-1, 1))
```

```python
linear_15 = linear_model.LinearRegression()
linear_15.fit(x_15, y)
(linear_15.coef_, linear_15.intercept_)
```

```python
poly_3 = PolynomialFeatures(degree = 3)
x_3 = poly_3.fit_transform(x.reshape(-1, 1))
```

```python
linear_3 = linear_model.LinearRegression()
linear_3.fit(x_3, y)
(linear_3.coef_, linear_3.intercept_)
```

```python
poly_2 = PolynomialFeatures(degree = 2)
x_2 = poly_2.fit_transform(x.reshape(-1, 1))
```

```python
linear_2 = linear_model.LinearRegression()
linear_2.fit(x_2, y)
(linear_2.coef_, linear_2.intercept_)
```

## Plot all the results for each polynomial.

```python
plt.scatter(x, y)
plt.plot(x, linear_15.predict(x_15) + linear_15.intercept_, c = 'b')
```

```python
plt.scatter(x, y)
plt.plot(x, linear_3.predict(x_3) + linear_3.intercept_, c = 'r')
```

```python
plt.scatter(x, y)
plt.plot(x, linear_2.predict(x_2) + linear_2.intercept_, c = 'g')
```
